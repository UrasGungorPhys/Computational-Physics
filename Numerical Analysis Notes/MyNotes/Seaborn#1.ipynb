```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('01_dataexport_20201008T180753.csv', skiprows=9)
df = df.rename(columns={"Ankara Temperature [2 m elevation corrected]":"Temperature",
         "Ankara Relative Humidity [2 m]":"Relative Humidity", "timestamp":"Timestamp",
         "Ankara Precipitation Total":"Precipitation Total"})
df.set_index("Timestamp")
df
```

```python
sns.set_style('dark')  # By far the most important command
plt1 = sns.relplot(data=df, x="Temperature", y="Relative Humidity")
```

```python
# not satissfied, need to go darker.
sns.set(rc={'axes.facecolor':'dimgray', 'figure.facecolor':'dimgray'})
plt1 = sns.relplot(data=df, x="Temperature", y="Relative Humidity", 
                   hue="Temperature", size="Relative Humidity")
```
