$$

    c) \,x = x + 1 -\frac{1}{a}x^2\\\,\\
    d) \,x = \frac{1}{2}\left(x+\frac{a}{x}\right)\\

```python
def f1(x,a):
    return x**2 + x - a

def f2(x,a):
    return a/x

def f3(x,a):
    return x + 1 -x**2/a

def f4(x,a):
    return 1/2*(x+a/x)
```

```python
# Inputs:

x0 = 3
a_vals = [1, 4, 9, 16, 25]  # Trying it out for different squares.
break_margin = 10e-5
max_iter = 100  # the number of iterations to give up on 

# --------------------------------------------------------------------------

computation = 1  # to make things pretty...
for a in a_vals:
    expected_result = np.sqrt(a)
    
    print("\n\n-----------------------------------------------------------------------------------------\n")
    print(f"\tCOMPUTATION {computation}, a = {a}:\n")

    # Part a:
    print(f"\n\t\tPART A (x = x^2 + x - a), a = {a}:\n\n")
    xi = x0
    xip1 = 0
    iterations = 0
    
    try:
        while True:

            xip1 = f1(xi,a)
            iterations +=1

            if xip1 > x0 * 10e7:
                print(f"\tSTOPPING on iteration {iterations}, explosion is imminent on the values of Part a!\n\t Last two iterations---> x{iterations-1} = {xi}, x{iterations} = {xip1}")
                break

            if np.abs(xip1 - xi) < break_margin:
                print(f"\tSUCCESS: Part A has achieved value x = {xip1}, in {iterations} iterations\n")
                print(f"\t The expected result was: {expected_result}, the deviation is {np.abs(np.abs(expected_result)-np.abs(xip1))},")
                
                if 0.0 < np.abs(np.abs(expected_result)-np.abs(xip1)) < break_margin:
                    print(f"\twhich is within the break margin {break_margin}", sep="")
                elif np.abs(np.abs(expected_result)-np.abs(xip1)) == 0:
                    print(f"\twhich is an exact result!", sep="")
                else:
                    print(f"\twhich has crept out of the break margin {break_margin}", sep="")

                break

            elif iterations >= max_iter:
                print(f"\tPart a has failed to converge after {iterations} iterations.")
                print(f"\t The value on the final two iterations were x{iterations-1} = {xi}, x{iterations} = {xip1}")
                break

            else:
                xi = xip1
                
    except OverflowError:
        print(f"\tFor a = {a} on Part A, the value seems to have diverged out of bit range, the program has self destruct...\n\t (OverflowError 34)")


    # Part b:
    print(f"\n\n\n\t\tPART B (x = a/x), a = {a}:\n")
    xi = x0
    xip1 = 0
    iterations = 0
    
    try:
        while True:

            xip1 = f2(xi,a)
            iterations +=1

            if xip1 > x0 * 10e7:
                print(f"\tSTOPPING on iteration {iterations}, explosion is imminent on the values of Part b!\n\t Last two iterations---> x{iterations-1} = {xi}, x{iterations} = {xip1}")
                break

            if np.abs(xip1 - xi) < break_margin:
                print(f"\tSUCCESS: Part B has achieved value x = {xip1}, in {iterations} iterations\n")
                print(f"\t The expected result was: {expected_result}, the deviation is {np.abs(np.abs(expected_result)-np.abs(xip1))},")
                
                if 0.0 < np.abs(np.abs(expected_result)-np.abs(xip1)) < break_margin:
                    print(f"\twhich is within the break margin {break_margin}", sep="")
                elif np.abs(np.abs(expected_result)-np.abs(xip1)) == 0:
                    print(f"\twhich is an exact result!", sep="")
                else:
                    print(f"\twhich has crept out of the break margin {break_margin}", sep="")

                break

            elif iterations >= max_iter:
                print(f"\tERROR: Part b has failed to converge after {iterations} iterations.")
                print(f"\t The value on the final two iterations were x_{iterations-1} = {xi}, x_{iterations} = {xip1}")
                break

            else:
                xi = xip1
                
    except OverflowError:
        print(f"\tFor a = {a} on Part B, the value seems to have diverged out of bit range, the program has self destruct...\n\t (OverflowError 34)")


    # Part c:
    print(f"\n\n\n\t\tPART C (x = x + 1 -x**2/a), a = {a}:\n")
    xi = x0
    xip1 = 0
    iterations = 0

    try:
        while True:

            xip1 = f3(xi,a)
            iterations +=1

            if xip1 > x0 * 10e7:
                print(f"\tSTOPPING on iteration {iterations}, explosion is imminent on the values of Part c!\n\t Last two iterations---> x{iterations-1} = {xi}, x{iterations} = {xip1}")
                break

            if np.abs(xip1 - xi) < break_margin:
                print(f"\tSUCCESS: Part c has achieved value x = {xip1}, in {iterations} iterations\n")
                print(f" \t The expected result was: {expected_result}, the deviation is {np.abs(np.abs(expected_result)-np.abs(xip1))},")
                
                if 0.0 < np.abs(np.abs(expected_result)-np.abs(xip1)) < break_margin:
                    print(f"\twhich is within the break margin {break_margin}", sep="")   
                elif np.abs(np.abs(expected_result)-np.abs(xip1)) == 0:
                    print(f"\twhich is an exact result!", sep="")
                else:
                    print(f"\twhich has crept out of the break margin {break_margin}", sep="")

                break

            elif iterations >= max_iter:
                print(f"\tPart c has failed to converge after {iterations} iterations.")
                print(f"\t The value on the final two iterations were x_{iterations-1} = {xi}, x_{iterations} = {xip1}")
                break

            else:
                xi = xip1
    
    except OverflowError:
        print(f"\tFor a = {a} on Part C, the value seems to have diverged out of bit range, the program has self destruct...\n\t (OverflowError 34)")

    # Part d:
    print(f"\n\n\n\t\tPART D (x = 1/2*(x+a/x)), a = {a}:\n")
    xi = x0
    xip1 = 0
    iterations = 0

    try:
        while True:

            xip1 = f4(xi,a)
            iterations +=1

            if xip1 > x0 * 10e7:
                print(f"\tSTOPPING on iteration {iterations}, explosion is imminent on the values of Part d!\n\t Last two iterations---> x{iterations-1} = {xi}, x{iterations} = {xip1}")
                break

            if np.abs(xip1 - xi) < break_margin:
                print(f"\tSUCCESS: Part d has achieved value x = {xip1}, in {iterations} iterations\n")
                print(f" \t The expected result was: {expected_result}, the deviation is {np.abs(np.abs(expected_result)-np.abs(xip1))},")
                
                if 0.0 < np.abs(np.abs(expected_result)-np.abs(xip1)) < break_margin:
                    print(f"\twhich is within the break margin {break_margin}", sep="")
                elif np.abs(np.abs(expected_result)-np.abs(xip1)) == 0:
                    print(f"\twhich is an exact result!", sep="")
                else:
                    print(f"\twhich has crept out of the break margin {break_margin}", sep="")

                break


            elif iterations >= max_iter:
                print(f"\tPart d has failed to converge after {iterations} iterations.")
                print(f"\t The value on the final two iterations were x_{iterations-1} = {xi}, x_{iterations} = {xip1}")
                break

            else:
                xi = xip1
                
    except OverflowError:
        print(f"\tFor a = {a} on Part D, the value seems to have diverged out of bit range, the program has self destruct...\n\t (OverflowError 34)")
    
    computation+=1
```
