## Question 1, via Euler's Method:

$$
    -\alpha f = \beta \frac{d^2 f}{dt^2}, \,\, f(0)=0,\,\, \dot{f}=\gamma
$$

$$\dot{f}_{i+1} = \dot{f}_i + \ddot{f_i}h$$
    
    
$$f_{i+1} = f_i + \dot{f}h$$

```python
a = 2  # value for alpha
b = 4  # value for beta
c = 3  # value for gamma
N1 = 1000

def fdd(f, a, b):  # f double dot for f, alpha (a), beta (b)
    return -a*f/b


tvals = np.linspace(0,5,N1)
h = tvals[1] - tvals[0]

fs = np.array([0])  # array for f, including f(0) = 0
fds = np.array([c])  # array for fdot, including fdot(0) = gamma (c)

for t in tvals[1:]:
    fdi = fds[-1] + fdd(fs[-1], a, b)*h
    fi = fs[-1] + fds[-1]*h

    fds = np.append(fds, fdi)
    fs = np.append(fs, fi)

# Part B - Plotting:

plt.plot(tvals, fds, label="fdot")
plt.plot(tvals, fs, label="f")
plt.title("Euler's Method")
plt.legend()
plt.show()

# Part C:

# first differentiation:

# first value:
fdot0 = (fs[1]-fs[0]) / h


reverse_fdots = np.array([fdot0])
for i in range(2, len(fs)-1):  # other values, up to the last
    fdoti = (fs[i+1] - fs[i-1])/(2*h)
    reverse_fdots = np.append(reverse_fdots, fdoti)

# last value:
fdotf = (fs[-1] - fs[-2])/h
reverse_fdots = np.append(reverse_fdots,fdotf)

# if we plot these values together and they fit on top of eachother, we can show our computation was true:
plt.plot(tvals[1:], reverse_fdots, label="rev_fdot")
plt.plot(tvals[1:],fds[1:], "--", label="fdot")
plt.title("Part C, fdot")
plt.legend()
plt.show()

# second differentiation:
fddot0 = (reverse_fdots[1] - reverse_fdots[0])
reverse_fddots = np.array([fddot0])

for i in range(2, len(reverse_fdots)-1):
    fddoti = (reverse_fdots[i+1] - reverse_fdots[i-1])/(2*h)
    reverse_fddots = np.append(reverse_fddots, fddoti)

fddotf = (reverse_fdots[-1] - reverse_fdots[-2])/h
reverse_fddots = np.append(reverse_fddots, fddotf)

# creating a similar plot:
fdds = fdd(fs, a,b)
plt.plot(tvals[2:], reverse_fddots, label="rev_fddot")
plt.plot(tvals[2:], fdds[2:], "--", label="fddot")
plt.title("Part C, fddot")
plt.legend()
plt.show()

```

## Question 3:

```python
def f3(x):
    return np.sin(x)/x

N3 = 3000
if N3%3 !=0:
    N3 = N3 - N3%3  # N3 has to be divisible by 3!
    
xvals = np.linspace(-np.pi, np.pi, N3+1)
h = xvals[1] - xvals[0]

yvals = f3(xvals)
print(sum(xvals==0))  # make sure we don't get a zero
plt.plot(xvals, yvals)
plt.show()
```

$$
    I = \frac{3h}{8}\left[f(x_0) + 3f(x_1) + 3f(x_2) + f(x_3)\right]
$$

```python
I = 0
loop_is = np.arange(0, N3, 3)
for i in loop_is:
    
    fx0i = f3(xvals[i])
    fx1i = f3(xvals[i+1])
    fx2i = f3(xvals[i+2])
    fx3i = f3(xvals[i+3])
    
    I_i = (3/8)*(fx0i + 3*fx1i + 3*fx2i + fx3i)*h
    I += I_i

print("Thus we get the result", I)
```

## Question 4:

```python
def f4(x):
    return np.sin(x)/x

N4 = 3000
xvals = np.linspace(-np.pi, np.pi, N4)
yvals = f4(xvals)
print(sum(xvals==0))  # make sure we don't get a zero
plt.plot(xvals, yvals)
plt.show()
```


(Rands within the curve)/(Total number of rands) = (Area of the curve)/(Total area)

(Area of the curve) = (Total area) x (Rands within)/(Total Rands)

```python
x0 = xvals[0]  # range for x
xf = xvals[-1]
delx = xf - x0

ymin = yvals.min()  # range for y
ymax = yvals.max()
dely = ymax - ymin

yrands = np.random.rand(N4) * dely + ymin  # random numbers within their ranges
xrands = np.random.rand(N4) * delx + x0

curve = f4(xrands)  # the curve for the random x values
plt.plot(xrands, curve, ".")

yrands_in_curve = yrands<curve  # filter for values within the curve

y_in = yrands[yrands_in_curve]  # values within the curve (for the plot)
x_for_y_in = xrands[yrands_in_curve]  # respective xvals (for the plot)
y_out = yrands[yrands>curve]  # values outside the curve (for the plot)
x_for_y_out = xrands[yrands>curve]  # respective xvals(for the plot)

plt.plot(x_for_y_in, y_in, ".")
plt.plot(x_for_y_out, y_out, ".")
plt.show()

A = dely*delx  # total area
n_in = sum(yrands_in_curve)  # number of random points within the curve
n_total = N4  # number of total random points

result4 = A * n_in/n_total
print(f"And the integral is: {result4}, for {N4} random numbers")

```
