```python
import numpy as np
import matplotlib.pyplot as plt

def yp(y):
    return y
```

```python
for N in [10, 20, 50, 100, 200]:
    
    ts = np.linspace(0,4,N)
    h = ts[1] - ts[0]
    analytical_y = np.exp(ts)
    y_vals = np.array([1])

    for t in ts[1:]:

        yi = y_vals[-1] + yp(y_vals[-1])*h
        y_vals = np.append(y_vals, yi)
        
    print(f'N = {N}')
    plt.plot(ts, y_vals)
    plt.plot(ts, analytical_y,"r--")
    plt.show()
    print('-'*50)
```

```python
def ypp(y):
    return y

for N in [10, 30, 50, 100, 200]:
    
    ts = np.linspace(0,4,N)
    h = ts[1] - ts[0]
    analytical_y = np.exp(ts)
    y_vals = np.array([1])
    yp_vals = np.array([1])

    for t in ts[1:]:

        ypi = yp_vals[-1] + ypp(yp_vals[-1])*h
        yi = y_vals[-1] + ypi*h
        yp_vals = np.append(yp_vals, ypi)
        y_vals = np.append(y_vals, yi)

    print(f'N = {N}')
    plt.plot(ts, y_vals)
    plt.plot(ts, yp_vals, "k")
    plt.plot(ts, analytical_y,"r--")
    plt.show()
    print('-'*50)

```
