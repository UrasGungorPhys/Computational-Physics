```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
def f(x):
    return np.sin(x)

N=100
xs = np.linspace(0,2*np.pi,N)
h = xs[1]-xs[0]
fps = np.array([(f(xs[1]) - f(xs[0]))/h])

for i in range(1,len(xs)-1):
    
    fp = (f(xs[i+1]) - f(xs[i-1]))/(2*h)
    fps = np.append(fps,fp)
    
fps = np.append(fps, (f(xs[N-1]) - f(xs[N-2]))/(h))
plt.plot(xs,f(xs))
plt.plot(xs,fps)
```
