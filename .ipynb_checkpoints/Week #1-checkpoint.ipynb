# Spring motion

$$F=-kx = m\ddot x\\
x = a_0 + a_1 t + a_2 t^2+a_3t^3+\dots\\
\ddot x = (2\times1)a_2+(3\times2)a_3 t+(4\times3)a_4t^2\dots\\
((2\times1)a_2+(3\times2)a_3 t+(4\times3)a_4t^2\dots)=-\tfrac{k}{m}(a_0 + a_1 t + a_2 t^2+a_3t^3+\dots)$$


$$2a_2 = -\tfrac{k}{m}a_0\\
(3\times2)a_3 =-\tfrac{k}{m}a_1\\
(4\times3)a_4 =-\tfrac{k}{m}a_2\\
\vdots$$

Initial conditions:
* displacement :$ x(t=0) = x_0$
* velocity : $\dot x (t=0) = v_0 = 0$

$$a_0 = x_0\\
a_1 = 0 \rightarrow a_{2n+1} = 0$$

```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
list(range(0,10,2))
```

```python
t = np.linspace(0,1,50)
k = 1 # N/m
x0=10E-2 #m
m = 1E-2 # kg

omega = np.sqrt(k/m)

def x_a(t):
    # Analytical solution
    return x0*np.cos(omega*t)
def x_t(t,nn=100):
    # "Taylor" expansion (power series expansion)
    a_nm2 = x0
    x = a_nm2
    for n in range(2,nn,2):
        a_n = -1/(n*(n-1))*k/m*a_nm2
        #print(n,a_n)
        x += a_n * t**n
        a_nm2 = a_n
    return x

def x_t2(t,nn=500):
    x = x0
    for n in range(1,nn):
        x += (-1)**n/np.math.factorial(2*n)*(k/m)**n*x0*t**(2*n)
    return x
```

```python
plt.plot(t,x_a(t),"b-")
start = time.time()
plt.plot(t,x_t(t,200),"r--")
end = time.time()
print(end - start)

start = time.time()
plt.plot(t,x_t2(t,100),"g--")
end = time.time()
print(end - start)

plt.show()
```

3) $\dydx{y}{x}=y,\,y(0)=1$

$\begin{align*}\dydx{y}{x}=y\rightarrow\frac{\diff y}{y}&=\diff{x} \\
\ln{y}+C &= x\\
\cancel {e}^{\cancel{\ln} y} + e^C &= e^x\\
y&=e^x + C\\
y(0) &= 1 + C = 1\rightarrow C=0\Rightarrow\boxed{y=e^x}
\end{align*}$

Using Taylor's expansion:

$$y = a_0+a_1x+a_2x^2+a_3x^3+a_4x^4+\dots = \sum_{n=0}^{\infty}{a_n x^n}$$
$$y'=a_1+2a_2x+3a_3x^2+4a_4x^3+\dots = \sum_{n=0}^{\infty}{n a_n x^{n-1}}$$

$$\begin{gather*}
y'=y:\\
a_1=a_0=1\;\text{(initial condition)}\\
2a_2=a_1=1\rightarrow a_2=\tfrac{1}{2}
3a_3=a_2=\tfrac{1}{2}\rightarrow a_2=\tfrac{1}{3!}\\
4a_4=a_3=\tfrac{1}{3!}\rightarrow a_3=\tfrac{1}{4!}\\
\vdots\\
a_n=\tfrac{1}{n!}
\end{gather*}$$

$$y=\sum_{n=0}^{\infty}{\frac{1}{n!} x^n}\equiv e^x$$

## Charging Capacitor
(FIZ138_LN_07, p9)
$$R\dydx{q}{t}+\frac{q}{C}=\mathcal{E},\;q(t=0)=0$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

[R*q'(t)+q(t)/C=A](https://www.wolframalpha.com/input?i=R*q%27%28t%29%2Bq%28t%29%2FC%3DA)

$$C\mathcal{E} + C_1e^{-t/{RC}}$$

$q(t=0)=0\rightarrow C_1 = -C\mathcal{E}$

$$\rightarrow\boxed{ q(t)=C\mathcal{E}\left(1-e^{-t/{RC}}\right)}$$

```python
def q_a(t):
    return C*Eps*(1-np.exp(-t/(R*C)))
```

$$q=a_0+a_1t+a_2t^2+\dots\\
q'=a_1+2a_2t+3a_3t^2+\dots$$

$$q(t = 0) = 0\rightarrow a_0 = 0$$

$$R(a_1+2a_2t+3a_3t^2+\dots)+\frac{(a_0+a_1t+a_2t^2+\dots)}{C}=\mathcal{E}\\
RCa_1 = \mathcal{E}C\rightarrow a_1 = \frac{\mathcal{E}}{R}\\
RC2a_2 + a_1 = 0\rightarrow a_2 = -\frac{a_1}{2RC}\\
RC3a_3 + a_2 = 0\rightarrow a_3 = -\frac{a_2}{3RC}\\
\vdots\\
a_n = -\frac{a_{n-1}}{nRC}
$$


```python
def q_n(t):
    a_pre = Eps/R
    q = a_pre * t
    for n in np.arange(2,25):
        a = -a_pre / (n*R*C)
        q += a * t**n
        a_pre = a
    return q
```

```python
C=10E-3 
R=50
Eps = 50
t=np.linspace(0,4,100)
```

```python
plt.plot(t,q_a(t),t,q_n(t),"--")
plt.show()
```

## Trevin formal/resmi tanm
Hatrlarsanz, matematiksel formasyonda trev u ekilde tanmlanmakta idi: 
$$\lim_{\Delta x\rightarrow 0}{\left[\frac{f(x+\Delta x)-f(x)}{\Delta x}\right]} = \frac{\text{d}f}{\text{d}x}$$

Halk dilinde ise (onunla ilgili karikatr koyamyorum, biraz ayp ;), "ben imdi unu biraz arttrrsam, bu ne kadar artar?" sorusunun cevab olarak tanmlyoruz trevi.

Yukardaki (formal olan) tanm kodda ifade edip, $x=\pi/3$ noktasnda $\sin(x)$ fonksiyonun trevini bulalm:

```python
import numpy as np

def f(x):
    return np.sin(x)

x = np.pi/3

delta_x = 1E-6

turev_x = (f(x+delta_x) - f(x)) / delta_x

gercek_deger = np.cos(x)

print("Hesaplanan deer: {:.15f}".format(turev_x))
print("    Gerek Deer: {:.15f}".format(gercek_deger))
print("            Hata: {:.15f}".format(np.abs(turev_x-gercek_deger)))
```

Tabii ki, aral ne kadar kk tutarsak (yukarda $10^{-6}$ aldk bu arada, _1E-6_ o demek), gerek deere de o kadar yaklaacamz bariz, burada o faktr tekrardan deerlendirmeyeceiz ama u faktre bir bakalm (bakalm koddaki fark fark edebilecek misiniz?)

```python
import numpy as np

def f(x):
    return np.sin(x)

x = np.pi/3

delta_x = 1E-6

turev_x = (f(x) - f(x-delta_x)) / delta_x

gercek_deger = np.cos(x)

print("Hesaplanan deer: {:.15f}".format(turev_x))
print("    Gerek Deer: {:.15f}".format(gercek_deger))
print("            Hata: {:.15f}".format(np.abs(turev_x-gercek_deger)))
```

Yapm olduumuz ey u: trev madem ki fonksiyonun verilen noktasnda fonksiyona teet geen dorunun eimi, doruyu tanmlamak iin de iki tane noktamz olmas gerekiyor, ite o iki noktay nasl setiimiz bize kalm: ilk uygulamada, doruyu izmek iin verilen nokta ile ondan $\Delta x$ kadar ilerisindeki noktay kullanrken; ikinci uygulamada verilen nokta ile onun $\Delta x$ kadar gerisindeki noktay kullandk. 

Grafiksel olarak u iki dorunun eimlerini aldk yani:

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x)

x = np.pi/3

delta_x = 1E-6

t = np.concatenate((np.linspace(0.7,x,5),np.linspace(x,1.3,5)))
y = f(t)

turev_0 = (f(x+delta_x)-f(x)) / delta_x
turev_1 = (f(x) - f(x-delta_x)) / delta_x

delta_y0 = f(x)-turev_0*x
delta_y1 = f(x)-turev_1*x


fig,ax = plt.subplots(2,1)

ax[0].plot(t,y,"*-r")
ax[0].plot(x,f(x),"o")
ax[0].plot(t,turev_0*t+delta_y0,"k--")

ax[1].plot(t,y,"*-r")
ax[1].plot(x,f(x),"o")
ax[1].plot(t,turev_1*t+delta_y1,"k--")
ax[0].title.set_text("Bunlar byle ayr ayr izdik ama pek bir fark anlalmyor 8P")
plt.show()
```

Bu durumda da, aklmza, integralde de gelen, _iki yaklamn ortas_ geliyor, yani, trevi seili noktann $\Delta x$ kadar ilerisi ya da gerisi ile hesaplamak yerine, hem ilerideki hem de gerideki noktalar iin iine katp, u ekilde hesaplamak:

$$f'(x) = \frac{f(x+\Delta x) - f(x-\Delta x)}{2\Delta x}$$

(araln $\Delta x$'den $2\Delta x$'e ktna dikkat edin..)

## Otomatik yapan yok mu bunun? (`scipy.misc.derivative()`)
Doal olarak, bu trev iini de yapan prosedrler mevcut, rnein: [`scipy.misc.derivative`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.misc.derivative.html):

```python
import numpy as np
from scipy.misc import derivative

def f(x):
    return np.sin(x)

x = np.pi/3
df_dx = derivative(f,x,dx=1E-6)
gercek_deger = np.cos(x)

print("           Hesaplanan Deer: {:.15f}".format(df_dx))
print("               Gerek Deer: {:.15f}".format(gercek_deger))
print("                       Hata: {:.15f}".format(np.abs(df_dx-gercek_deger)))
```

## Five-point method
$$\{x-2h, x-h, x, x+h, x+2h\}$$

$$f'(x) = \frac{-f(x + 2h) + 8 f(x + h) - 8 f(x - h) + f(x - 2h)}{12h} + \frac{h^4}{30} f^{(5)}(c)\\(c \in [x - 2h, x + 2h])$$

$$f(x \pm h) = f(x) \pm h f'(x) + \frac{h^2}{2}f''(x) \pm \frac{h^3}{6} f^{(3)}(x) + O_{1\pm}(h^4). \qquad (E_{1\pm}).$$

$$(E_{1+})-(E_{1-})$$

$$f(x+h) - f(x-h) = 2hf'(x) + \frac{h^3}{3}f^{(3)}(x) + O_1(h^4). \qquad (E_1).$$


Similarly, we have

$$f(x \pm 2h) = f(x) \pm 2h f'(x) + \frac{4h^2}{2!} f''(x) \pm \frac{8h^3}{3!} f^{(3)}(x) + O_{2\pm}(h^4). \qquad (E_{2\pm})$$

and $$(E_{2+})-(E_{2-})$$

$$f(x+2h) - f(x-2h) = 4hf'(x) + \frac{8h^3}{3}f^{(3)}(x) + O_2(h^4). \qquad (E_2).$$

In order to eliminate the terms of &fnof;<sup>&nbsp;(3)</sup>(''x''), calculate 8&nbsp;&times;&nbsp;(''E''<sub>1</sub>)&nbsp;&minus;&nbsp;(''E''<sub>2</sub>)

$$8f(x+h) - 8f(x-h) - f(x+2h) + f(x-2h) = 12h f'(x) + O(h^4)$$
