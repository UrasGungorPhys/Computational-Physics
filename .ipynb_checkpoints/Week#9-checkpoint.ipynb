```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as spi
np.random.seed(468)
```

```python
def gaussian(x, mu, sigma):
    return np.exp(-(x-mu)**2/(2*sigma**2))

def gauss0(x):
    return gaussian(x, 5, 1)


def monte_carlo_integration(function,x0,x1, N):
    
    xvals = np.linspace(x0, x1, N)
    xpoints = np.random.rand(N)*(x1 - x0) + x0
    curve = function(xpoints)
    y0 = np.amin(curve)
    y1 = np.amax(curve)
    ypoints = np.random.rand(N)*(y1 - y0) + y0
    plt.plot(xvals, function(xvals))
    plt.plot(xpoints, ypoints, '.')
    
    A = np.abs(np.abs(x1 - x0)*np.abs(y1 - y0))
    
    
    py = ypoints*(ypoints>0)
    ny = ypoints*(ypoints<0)
    
    py_in_curve = (py<curve) & (py!=0)
    ny_in_curve = (ny>curve) & (ny!=0)
    plt.plot(xpoints[py_in_curve], ypoints[py_in_curve],'.')
    
    S = A*np.abs(np.sum(py_in_curve)-np.sum(ny_in_curve))/N
    plt.show()
    
    return S


monte_carlo_integration(gauss0,0,10,10000)
```

```python
def f1(x):
    return np.exp(0.4*(x-0.4)**2-0.08*x**4)

xs = np.linspace(0,4,1000)
print(spi.quad(f1, -np.infty, np.infty))
print(monte_carlo_integration(f1, -3.5, 3.5, 10000))

```

```python
N = 1000
def gauss2(x):
    return gaussian(x, 5,1)
    

xvals = np.linspace(-5,5, N)
yvals = np.linspace(-5,5, N)

gaussx = gauss2(xvals) - 5
gaussy = gauss2(yvals) - 5

```

```python
xrands = np.random.rand(N)*10 -5
yrands = np.random.rand(N)*10 -5

plt.plot(xrands*(np.abs(xrands)<gaussx), yrands*(yrands<gaussy), 'o')

#2d gaussian in polar coordinates
```
