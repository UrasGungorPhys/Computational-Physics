# Question 1:

$$
    f(x) = x^2 - a, \,\, a>0 \\
$$




$$
    a) \,x = x^2 + x - a \\\,\\
    b) \,x = \frac{a}{x}\\\,\\
    c) \,x = x + 1 -\frac{1}{a}x^2\\\,\\
    d) \,x = \frac{1}{2}\left(x+\frac{a}{x}\right)\\
$$ 

```python
def f1(x,a):
    return x**2 + x - a

def f2(x,a):
    return a/x

def f3(x,a):
    return x + 1 -x**2/a

def f4(x,a):
    return 1/2*(x+a/x)
```

### The algorithm, without the prints, for readibility:

**Inputs:**

    x0 = 3
    a_vals = [1, 4, 9, 16, 25]  Trying it out for different squares.
    break_margin = 10e-5
    max_iter = 100  # the number of iterations to give up on 

**The loop for a values:**

    for a in a_vals:
        expected_result = np.sqrt(a)
        
   ***The code for Part A (which is then copy pasted for the different parts, with different functions)***
        
        xi = x0
        xip1 = 0
        iterations = 0
        
        try: (to prevent OverflowError crash)
        
            while True:

                xip1 = f1(xi,a)
                iterations +=1

                if xip1 > x0 * 10e7:
                 (break for divergence, at 10e7 we can very confidently say we have failed)
                    break

                if np.abs(xip1 - xi) < break_margin:
                    (break for successful result)

                    (but also check the deviation)
                    
                    if 0.0 < np.abs(np.abs(expected_result)-np.abs(xip1)) < break_margin:
                        (the deviation is within the margin)
                    elif np.abs(np.abs(expected_result)-np.abs(xip1)) == 0:
                        (the result it exact)
                    else:
                        (the deviation is larger than expected)

                    break

                elif iterations >= max_iter:
                    (break for many iterations)
                    break

                else:
                    xi = xip1 (prep the values for next loop)

        except OverflowError:
            (insert custom error message here)


```python
# Inputs:

x0 = 3
a_vals = [1, 4, 9, 16, 25, 36]  # Trying it out for different squares.
break_margin = 10e-5
max_iter = 100  # the number of iterations to give up on 

# --------------------------------------------------------------------------

computation = 1  # to make things pretty...
for a in a_vals:
    expected_result = np.sqrt(a)
    
    print("\n\n-----------------------------------------------------------------------------------------\n")
    print(f"\tCOMPUTATION {computation}, a = {a}:\n")

    # Part a:
    print(f"\n\t\tPART A (x = x^2 + x - a), a = {a}:\n\n")
    xi = x0
    xip1 = 0
    iterations = 0
    
    try:
        while True:

            xip1 = f1(xi,a)
            iterations +=1

            if xip1 > x0 * 10e7:
                print(f"\tSTOPPING on iteration {iterations}, explosion is imminent on the values of Part a!\n\t Last two iterations---> x{iterations-1} = {xi}, x{iterations} = {xip1}")
                break

            if np.abs(xip1 - xi) < break_margin:
                print(f"\tSUCCESS: Part A has achieved value x = {xip1}, in {iterations} iterations\n")
                print(f"\t The expected result was: {expected_result}, the deviation is {np.abs(np.abs(expected_result)-np.abs(xip1))},")
                
                if 0.0 < np.abs(np.abs(expected_result)-np.abs(xip1)) < break_margin:
                    print(f"\twhich is within the break margin {break_margin}", sep="")
                elif np.abs(np.abs(expected_result)-np.abs(xip1)) == 0:
                    print(f"\twhich is an exact result!", sep="")
                else:
                    print(f"\twhich has crept out of the break margin {break_margin}", sep="")

                break

            elif iterations >= max_iter:
                print(f"\tPart a has failed to converge after {iterations} iterations.")
                print(f"\t The value on the final two iterations were x{iterations-1} = {xi}, x{iterations} = {xip1}")
                break

            else:
                xi = xip1
                
    except OverflowError:
        print(f"\tFor a = {a} on Part A, the value seems to have diverged out of bit range, the program has self destruct...\n\t (OverflowError 34)")


    # Part b:
    print(f"\n\n\n\t\tPART B (x = a/x), a = {a}:\n")
    xi = x0
    xip1 = 0
    iterations = 0
    
    try:
        while True:

            xip1 = f2(xi,a)
            iterations +=1

            if xip1 > x0 * 10e7:
                print(f"\tSTOPPING on iteration {iterations}, explosion is imminent on the values of Part b!\n\t Last two iterations---> x{iterations-1} = {xi}, x{iterations} = {xip1}")
                break

            if np.abs(xip1 - xi) < break_margin:
                print(f"\tSUCCESS: Part B has achieved value x = {xip1}, in {iterations} iterations\n")
                print(f"\t The expected result was: {expected_result}, the deviation is {np.abs(np.abs(expected_result)-np.abs(xip1))},")
                
                if 0.0 < np.abs(np.abs(expected_result)-np.abs(xip1)) < break_margin:
                    print(f"\twhich is within the break margin {break_margin}", sep="")
                elif np.abs(np.abs(expected_result)-np.abs(xip1)) == 0:
                    print(f"\twhich is an exact result!", sep="")
                else:
                    print(f"\twhich has crept out of the break margin {break_margin}", sep="")

                break

            elif iterations >= max_iter:
                print(f"\tERROR: Part b has failed to converge after {iterations} iterations.")
                print(f"\t The value on the final two iterations were x_{iterations-1} = {xi}, x_{iterations} = {xip1}")
                break

            else:
                xi = xip1
                
    except OverflowError:
        print(f"\tFor a = {a} on Part B, the value seems to have diverged out of bit range, the program has self destruct...\n\t (OverflowError 34)")


    # Part c:
    print(f"\n\n\n\t\tPART C (x = x + 1 -x**2/a), a = {a}:\n")
    xi = x0
    xip1 = 0
    iterations = 0

    try:
        while True:

            xip1 = f3(xi,a)
            iterations +=1

            if xip1 > x0 * 10e7:
                print(f"\tSTOPPING on iteration {iterations}, explosion is imminent on the values of Part c!\n\t Last two iterations---> x{iterations-1} = {xi}, x{iterations} = {xip1}")
                break

            if np.abs(xip1 - xi) < break_margin:
                print(f"\tSUCCESS: Part c has achieved value x = {xip1}, in {iterations} iterations\n")
                print(f" \t The expected result was: {expected_result}, the deviation is {np.abs(np.abs(expected_result)-np.abs(xip1))},")
                
                if 0.0 < np.abs(np.abs(expected_result)-np.abs(xip1)) < break_margin:
                    print(f"\twhich is within the break margin {break_margin}", sep="")   
                elif np.abs(np.abs(expected_result)-np.abs(xip1)) == 0:
                    print(f"\twhich is an exact result!", sep="")
                else:
                    print(f"\twhich has crept out of the break margin {break_margin}", sep="")

                break

            elif iterations >= max_iter:
                print(f"\tPart c has failed to converge after {iterations} iterations.")
                print(f"\t The value on the final two iterations were x_{iterations-1} = {xi}, x_{iterations} = {xip1}")
                break

            else:
                xi = xip1
    
    except OverflowError:
        print(f"\tFor a = {a} on Part C, the value seems to have diverged out of bit range, the program has self destruct...\n\t (OverflowError 34)")

    # Part d:
    print(f"\n\n\n\t\tPART D (x = 1/2*(x+a/x)), a = {a}:\n")
    xi = x0
    xip1 = 0
    iterations = 0

    try:
        while True:

            xip1 = f4(xi,a)
            iterations +=1

            if xip1 > x0 * 10e7:
                print(f"\tSTOPPING on iteration {iterations}, explosion is imminent on the values of Part d!\n\t Last two iterations---> x{iterations-1} = {xi}, x{iterations} = {xip1}")
                break

            if np.abs(xip1 - xi) < break_margin:
                print(f"\tSUCCESS: Part d has achieved value x = {xip1}, in {iterations} iterations\n")
                print(f" \t The expected result was: {expected_result}, the deviation is {np.abs(np.abs(expected_result)-np.abs(xip1))},")
                
                if 0.0 < np.abs(np.abs(expected_result)-np.abs(xip1)) < break_margin:
                    print(f"\twhich is within the break margin {break_margin}", sep="")
                elif np.abs(np.abs(expected_result)-np.abs(xip1)) == 0:
                    print(f"\twhich is an exact result!", sep="")
                else:
                    print(f"\twhich has crept out of the break margin {break_margin}", sep="")

                break


            elif iterations >= max_iter:
                print(f"\tPart d has failed to converge after {iterations} iterations.")
                print(f"\t The value on the final two iterations were x_{iterations-1} = {xi}, x_{iterations} = {xip1}")
                break

            else:
                xi = xip1
                
    except OverflowError:
        print(f"\tFor a = {a} on Part D, the value seems to have diverged out of bit range, the program has self destruct...\n\t (OverflowError 34)")
    
    computation+=1
```

## Conclusions:

- **Part A** was successful twice, on a = 9 and a = 16, and diverged on the other 4 cases.
- **Part B** was successful once, on a = 9, and oscillated on the other 5 cases.
- **Part C** was successful 5 times, and had an OverflowError on a = 1.
- **Part D** was successful every single time!

 All of the parts were clearly expected to be successful on a=9, as their initial value was already the right answer, so that case doesn't really count.

 That means **Part B** was actually never successful, which is to be expected, as it just takes the 1/x of each input everytime, it should always oscillate between two values.
 
 **Part A** was then only successful once, on a=16, where for $x=3$ we get $x^2 + x - a = 9 + 3 - 16 = -4$ on the first iteration, which means Part A just got lucky on that value. That also kinda doesn't count, so Part A is also a complete failure.
 
**Part C** was then successful 4 times, pretty good. Let's print all the values to see what went wrong on a=1:

```python
a = 1
x0 = 3
break_margin = 10e-5
max_iter = 100
print(f"\n\n\n\t\tPART C (x = x + 1 -x**2/a), a = {a}:\n")
xi = x0
xip1 = 0
iterations = 0

try:
    while True:

        xip1 = f3(xi,a)
        iterations +=1
        print(f"\nIteration: {iterations}\nx_i = {xi}, x_i+1 = {xip1}")

        if xip1 > x0 * 10e7:
            print(f"\tSTOPPING on iteration {iterations}, explosion is imminent on the values of Part c!\n\t Last two iterations---> x{iterations-1} = {xi}, x{iterations} = {xip1}")
            break

        if np.abs(xip1 - xi) < break_margin:
            print(f"\tSUCCESS: Part c has achieved value x = {xip1}, in {iterations} iterations\n")
            print(f" \t The expected result was: {expected_result}, the deviation is {np.abs(np.abs(expected_result)-np.abs(xip1))},")

            if 0.0 < np.abs(np.abs(expected_result)-np.abs(xip1)) < break_margin:
                print(f"\twhich is within the break margin {break_margin}", sep="")   
            elif np.abs(np.abs(expected_result)-np.abs(xip1)) == 0:
                print(f"\twhich is an exact result!", sep="")
            else:
                print(f"\twhich has crept out of the break margin {break_margin}", sep="")

            break

        elif iterations >= max_iter:
            print(f"\tPart c has failed to converge after {iterations} iterations.")
            print(f"\t The value on the final two iterations were x_{iterations-1} = {xi}, x_{iterations} = {xip1}")
            break

        else:
            xi = xip1

except OverflowError:
    print(f"\n\tFor a = {a} on Part C, the value seems to have diverged out of bit range, the program has self destruct...\n\t (OverflowError 34)")
```

**Moving on then...** 

**Part D** has been successful on every single case. 

 $x = \frac{1}{2}\left(x+\frac{a}{x}\right)$ doesn't seem to have any of the issues the other functions have. For $x\neq0$ of course, there is no reason for it to oscillate or diverge. While all the other functions somewhat rely on the value of $a$ to get them to converge, this one also has the handy $\frac{1}{2}$ coefficient that acts onto $x+\frac{a}{x}$, which gives it a greater chance of making the result smaller on every iteration, making it less likely to diverge.

For the next iteration to be larger than the previous one:

$$
    x_i = \frac{1}{2} \left( x_{i-1} + \frac{a}{x_{i-1}} \right)
$$

For $x_i > x_{i-1}$: 

$$
   \frac{1}{2} \left( x_{i-1} + \frac{a}{x_{i-1}} \right) > x_{i-1}\\
$$

$$
\rightarrow \:\frac{a}{2x_{i-1}} > \frac{1}{2} x_{i-1}\\
$$

$$
    \fbox{$a > x_{i-1}^2$}
$$



After this happens, since $x_i > x_{i-1}$, the value will increase on every iteration until $a=x_i$ is satisfied, so everything looks good.

```python
a = 100000
x0 = 3
break_margin = 10e-5
max_iter = 100
xi = x0
xip1 = 0
iterations = 0
expected_result = np.sqrt(a)
print(f"\n\n\n\t a = {a}, x_0 = {x0}:\n")
try:
    while True:

        xip1 = f4(xi,a)
        iterations +=1

        if xip1 > x0 * 10e7:
            print(f"\tSTOPPING on iteration {iterations}, explosion is imminent on the values of Part d!\n\t Last two iterations---> x{iterations-1} = {xi}, x{iterations} = {xip1}")
            break

        if np.abs(xip1 - xi) < break_margin:
            print(f"\tSUCCESS: Part d has achieved value x = {xip1}, in {iterations} iterations\n")
            print(f" \t The expected result was: {expected_result}, the deviation is {np.abs(np.abs(expected_result)-np.abs(xip1))},")

            if 0.0 < np.abs(np.abs(expected_result)-np.abs(xip1)) < break_margin:
                print(f"\twhich is within the break margin {break_margin}", sep="")
            elif np.abs(np.abs(expected_result)-np.abs(xip1)) == 0:
                print(f"\twhich is an exact result!", sep="")
            else:
                print(f"\twhich has crept out of the break margin {break_margin}", sep="")

            break


        elif iterations >= max_iter:
            print(f"\tPart d has failed to converge after {iterations} iterations.")
            print(f"\t The value on the final two iterations were x_{iterations-1} = {xi}, x_{iterations} = {xip1}")
            break

        else:
            xi = xip1

except OverflowError:
    print(f"\tFor a = {a} on Part D, the value seems to have diverged out of bit range, the program has self destruct...\n\t (OverflowError 34)")
    
    
    
a = 9
x0 = 100000
break_margin = 10e-5
max_iter = 100
xi = x0
xip1 = 0
iterations = 0
expected_result = np.sqrt(a)
print(f"\n\n\n\t a = {a}, x_0 = {x0}:\n")

try:
    while True:

        xip1 = f4(xi,a)
        iterations +=1

        if xip1 > x0 * 10e7:
            print(f"\tSTOPPING on iteration {iterations}, explosion is imminent on the values of Part d!\n\t Last two iterations---> x{iterations-1} = {xi}, x{iterations} = {xip1}")
            break

        if np.abs(xip1 - xi) < break_margin:
            print(f"\tSUCCESS: Part d has achieved value x = {xip1}, in {iterations} iterations\n")
            print(f" \t The expected result was: {expected_result}, the deviation is {np.abs(np.abs(expected_result)-np.abs(xip1))},")

            if 0.0 < np.abs(np.abs(expected_result)-np.abs(xip1)) < break_margin:
                print(f"\twhich is within the break margin {break_margin}", sep="")
            elif np.abs(np.abs(expected_result)-np.abs(xip1)) == 0:
                print(f"\twhich is an exact result!", sep="")
            else:
                print(f"\twhich has crept out of the break margin {break_margin}", sep="")

            break


        elif iterations >= max_iter:
            print(f"\tPart d has failed to converge after {iterations} iterations.")
            print(f"\t The value on the final two iterations were x_{iterations-1} = {xi}, x_{iterations} = {xip1}")
            break

        else:
            xi = xip1

except OverflowError:
    print(f"\tFor a = {a} on Part D, the value seems to have diverged out of bit range, the program has self destruct...\n\t (OverflowError 34)")
    
    
    
```

# Question 2:

$$
    f(x) = -x^5 + 5x^4 + \cos{(0.2x)}
$$

```python
def f(x):
    return -x**5 + 5*x**4 + np.cos(0.2*x)

x_vals = np.linspace(0,5.5,200)
plot = plt.plot(x_vals, f(x_vals))
```

## For  $|f(c)| < 10^{-5}$:

```python
eps = 1e-5
a = x_vals[0]
b = x_vals[-1]
iterations = 0
max_iterations = 100

while True:
    c = (a+b)/2
    if np.abs(f(c)) < eps:
        print(f"x = {c} +- {eps} is a root!\n")
        print(f"f(x-{eps}) = {f(c-eps)}\nf(x) = {f(c)}\nf(x+{eps}) = {f(c+eps)}")
        break
        
    if np.sign(f(a)*f(c)) == -1:
        b = c
        
    else:
        a = c
    # there is no other option here, since we have chosen a and b as the maximum and minimum x values.
```

# For $\Delta x < 10^{-5}$:

```python
eps = 1e-5
a = x_vals[0]
b = x_vals[-1]
iterations = 0
max_iterations = 100

while True:
    c = (a+b)/2
    if np.abs(c) - np.abs(a) < 2*eps:
        print(f"x = {c} +- {eps} is a root!\n")
        print(f"f(x-{eps}) = {f(c-eps)}\nf(x) = {f(c)}\nf(x+{eps}) = {f(c+eps)}")
        break
        
    if np.sign(f(a)*f(c)) == -1:
        b = c
        
    else:
        a = c
```

Looks like the $|f(c)| < 10^{-5}$ approach performed better!
