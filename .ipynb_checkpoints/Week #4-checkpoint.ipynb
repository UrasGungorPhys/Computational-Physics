    I = \frac{3h}{8}[f(x_0) + 3f(x_1) + 3f(x_2) + f(x_3)] + \frac{3h}{8}[f(x_4) + 3f(x_5) + 3f(x_6) + f(x_7)] + \frac{3h}{8}[f(x_8) + 3f(x_9) + 3f(x_{10}) + f(x_{11})] + ...

```python
n = 7
x = np.linspace(0,0.8,n)
h = x[1] - x[0]

f_oddi = 0
f3ni = 0
f_eveni = 0

for i in range(1,n-1,2):
    if i%3 == 0:
        continue
#     print(f"i1 = {i}")
    f_oddi += p(x[i])

for i in range(2,n-1,2):
    if i%3 == 0:
        continue
#     print(f"i2 = {i}")
    f_eveni += p(x[i])
    
for i in range(1,n-1):
    if i%3 == 0:
        f3ni+=p(x[i])
#         print(f"i3 = {i}")
    else:
        continue

    
    
int_s = (3*h/8)*(p(x[0]) + 3*f_oddi + 3*f_eveni + 2*f3ni + p(x[n-1]))

print(f"result = {int_s}, err = {(true_int - int_s)/true_int*100}%")
```
